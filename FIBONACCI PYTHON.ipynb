                                  FIBONACCI PYTHON EXECUTION WITH TIME MEASURE

```python
## El codigo para Fibonacci iterativo es el siguiente
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Como python no tiene tipos de variables solo basta hacer una funcion
import time

def fibonacci(n): 
    start = time.time()
    i = 1
    j = 0
 
    for k in range(n) :
        j = j + i
        i = j - i
    end = time.time()
    # print "El numero", n, "de fibonacci es: ", j
    return end - start
fibonacci(50)
```

                                   Impresion de grafica del tiempo de ejecucion

```python
nthfibo = []
times = []
#Hallaremos hasta el 600avo numero de fibonacci
for k in range(600) :
    nthfibo += [k]
    #medimos el tiempo
    times += [fibonacci(k)]
    #Y graficamos
plt.plot(nthfibo,times)
plt.title('Time to find the nth number of the fibonacci succesion')
plt.ylabel('Time')
plt.xlabel('Fibonacci No. x')
plt.legend('Time');
```

                                     Graficando varias veces la funcion

```python
MTimes = []
pruebas = 100
numero = []
#A cada numero de fibonacci se le realizaran 'pruebas' numero de evaluaciones 
#Se trato de usar la libreria Threads y Multiprocessing pero Quantopian lamentablemente no las soporta
for k in range(600):
    prom = 0
    numero += [k]
    for i in range(pruebas+1) :
        prom +=  fibonacci(k)
    MTimes += [(prom / pruebas)]
    # y se graficaran
plt.plot(numero,MTimes)
plt.title('Average Time to find the nth number of the fibonacci succesion')
plt.ylabel('Multiple Times')
plt.xlabel('Fibonacci No. x')
plt.legend('Time');
```
